## 將資料期間曾經產生虧損的公司剔除掉

- 使用的金融數據為 [datasets](https://classroom-hahow-learn-python-with-stock-data.s3.ap-northeast-1.amazonaws.com/datasets.zip) 中的 `gross_profits_and_operating_margins.json`
- 金融數據擷取來源：[Yahoo! Finance](https://github.com/ranaroussi/yfinance)

## 對於「虧損」的接受度

- 假如我們對於「虧損」的接受度比較寬鬆，意即我不想在這個條件上剔除掉太多公司，可以用「毛利（Gross profit）大於零」定義為沒有虧損。
- 假如我們對於「虧損」的接受度比較嚴苛，意即我希望在這個條件上盡可能剔除公司，可以用「淨利（Net income）大於零」定義為沒有虧損。

$$
\text{Gross profit} = \text{Revenue} - \text{Cost of goods sold} \\
\text{Net income} = \text{Gross profit} - \text{Operating expenses} -\\
\text{Depreciation and amortization} − \text{Taxes and other expenses}
$$

## 解題思路

- 首先將 gross_profits_and_operating_margins.json 上傳至 Google Colab 的工作目錄。
- 載入 `json` 模組。
- 使用 `with` 敘述搭配 `json.load()` 函數載入為 `list`
- 使用 `for` 迴圈走訪 200 個 `dict`
- 判斷 `gross_profit` 或 `net_income` 這兩個鍵所對應的多個值中是否有一個小於零（以 `net_income` 為例）。
- 將判斷結果加總，加總為 0 的紀錄為 `True`，反之紀錄為 `False`，儲存在一個新的 `dict` 之中。

````python
import json
import numpy as np

file_path = "gross_profits_and_operating_margins.json"
with open(file_path, "r") as f:
    gross_profits_and_operating_margins_json = json.load(f)
````

````python
is_profitable = list()
for data in gross_profits_and_operating_margins_json:
    ticker = data["ticker"]
    gross_profit = data["gross_profit"]
    net_income = data["net_income"]
    if net_income is None:
        dict_to_append = {
            "ticker": ticker,
            "is_profitable": False
        }
        is_profitable.append(dict_to_append)
    else:
        list_net_income_values = list(net_income.values())
        arr = np.array(list_net_income_values)
        arr_is_negative = arr < 0
        arr_sum = arr_is_negative.sum()
        dict_to_append = {
            "ticker": ticker,
            "is_profitable": not bool(arr_sum)
        }
        is_profitable.append(dict_to_append)
````
